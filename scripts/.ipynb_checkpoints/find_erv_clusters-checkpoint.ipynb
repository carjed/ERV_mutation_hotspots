```python
import csv
from collections import defaultdict
```

```python
outfile = "/mnt/norbert/data/bridges/bridges_ervs_3560_dist.txt"
outfh = open(outfile, 'w')

infile = "/mnt/norbert/data/bridges/bridges_ervs_3560.txt"
with open(infile, 'r') as f:
    reader = csv.DictReader(f, delimiter='\t')

    prevrow = {'CHR': '', 'ID': '', 'POS': '0'}
    last_site = {}
    
    site_count = 0
    for row in reader:

        ID = row['ID']
        chrom = row['CHR']
        if chrom == prevrow['CHR']:
            dist = abs(int(row['POS']) - int(prevrow['POS']))
            if ID in last_site:
                dist_ID = abs(int(row['POS']) - int(last_site[ID]['POS']))
            else:
                dist_ID = 0
                
            newrow = list(row.values()) + [str(dist), str(dist_ID)]
            outfh.write("\t".join(newrow) + "\n")
#             print(" ".join(newrow))
    
        site_count += 1        
        prevrow = row
        last_site[ID] = row
        
#         if site_count > 100:
#             break

outfh.close()
```
